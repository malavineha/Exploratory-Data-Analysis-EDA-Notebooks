## 1.3 Perform appropriate imputations to handle the missing values.


```python
# column 2-alpha code
df['2-alpha code'].unique()
```

```python
missing_rows = df[df['2-alpha code'].isnull()]
missing_rows
```

```python
missing_rows = df[df['Currency Unit'].isnull()]
missing_rows
```

```python
df['Region'].mode()
```

```python
df['Income Group'].mode()
```

```python
df['2-alpha code'] = df['2-alpha code'].fillna('Unknown')
df['Currency Unit'] = df['Currency Unit'].fillna(df['Currency Unit'].mode()[0])
df['Region'] = df['Region'].fillna(df['Region'].mode()[0])
df['Income Group'] = df['Income Group'].fillna(df['Income Group'].mode()[0])
df['WB-2 code']=df['WB-2 code'].fillna('Unknown')
```

```python
df['SNA price valuation'] = df['SNA price valuation'].fillna('Unknown')
df['Lending category'] = df['Lending category'].fillna('Unknown')
df['Other groups'] = df['Other groups'].fillna('Unknown')
df['System of National Accounts'] = df['System of National Accounts'].fillna('Unknown')
```

```python
df.drop(columns=['Alternative conversion factor','PPP survey year'],inplace=True)
```

```python
df['Balance of Payments Manual in use'] = df['Balance of Payments Manual in use'].fillna('Unknown')
df['External debt Reporting status'] = df['External debt Reporting status'].fillna('Unknown')
df['System of trade'] = df['System of trade'].fillna(df['System of trade'].mode()[0])
df['Government Accounting concept'] = df['Government Accounting concept'].fillna(df['Government Accounting concept'].mode()[0])
df['IMF data dissemination standard'] = df['IMF data dissemination standard'].fillna(df['IMF data dissemination standard'].mode()[0])
```

```python
df['National accounts base year']=df['National accounts base year'].fillna(method='ffill')
df['National accounts reference year']=df['National accounts reference year'].fillna(method='bfill')
```

```python
missing_rows = df[df['National accounts base year'].isnull()]
missing_rows
```

```python
df['Latest population census'] = df['Latest population census'].fillna(df['Latest population census'].mode()[0])
df['Latest household survey'] = df['Latest household survey'].fillna('Unknown')
df['Source of most recent Income and expenditure data'] = df['Source of most recent Income and expenditure data'].fillna('Unknown')
df['Vital registration complete'] = df['Vital registration complete'].fillna('Unknown')
df['Latest agricultural census'] = df['Latest agricultural census'].fillna(df['Latest agricultural census'].mode()[0])
```

```python
df['Latest industrial data'] = df['Latest industrial data'].fillna(df['Latest industrial data'].median())
df['Latest trade data'] = df['Latest trade data'].fillna(df['Latest trade data'].median())
```

```python
df['Balance of Payments Manual in use'] = df['Balance of Payments Manual in use'].fillna('Unknown')
```
