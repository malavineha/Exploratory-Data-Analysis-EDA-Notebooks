## Perform the One Hot Encoding separately on features *–degree_t*

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_excel('Placement_Dataset.xlsx')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['degree_t'].unique()
```

```python
df['hsc_s'].unique()
```

```python
ohe=OneHotEncoder()
```

```python
print(ohe)
```

### fit_transform method to convert the output to an array.

```python
ohe.fit_transform(df[['degree_t']]).toarray()
```

```python
feature_array=ohe.fit_transform(df[['degree_t']]).toarray()
```

```python
feature_labels=ohe.categories_
feature_labels
```

```python
feature_labels=np.array(feature_labels).ravel()
feature_labels
```

```python
df1=pd.DataFrame(feature_array,columns=feature_labels)
df1
```

```python
df2=pd.concat([df,df1],axis=1)
df2
```

##  Perform the One Hot Label separately on features –status

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_excel('Placement_Dataset.xlsx')
df.head(10)
```

```python
le=LabelEncoder()
```

```python
print(le)
```

```python
encoded_label=le.fit_transform(df[['status']])
encoded_label
```

```python
print("Original_categories",df[['status']])
print("Encoded_label",encoded_label)
```
